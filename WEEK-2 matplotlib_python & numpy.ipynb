Plotting using legend with 4 columns

```python
line1, = plt.plot(iris_data['PetalWidthCm'], label='PetalWidthCm',color='green',linewidth=1)
line2, = plt.plot(iris_data['PetalLengthCm'], label='PetalLengthCm',color='purple')
line3, = plt.plot(iris_data['SepalLengthCm'], label='SepalLengthCm',color='blue')
line4, = plt.plot(iris_data['SepalWidthCm'], label='SepalWidthCm',color='red')
plt.legend(handles=[line1, line2,line3,line4])
```

Bar Graphs

```python
data =iris_data.head()
```

```python
ax=data.plot(x="Id", y=['PetalWidthCm' ,'PetalLengthCm', 'SepalWidthCm','SepalLengthCm'],kind="bar", rot=0,figsize=(10,5),ylabel="Values")
```

```python
ax=data.plot(x="Id", y=['PetalWidthCm' ,'PetalLengthCm', 'SepalWidthCm','SepalLengthCm'],kind="bar", rot=0,figsize=(10,5),ylabel="Values",stacked='True')
```

SubPlots

```python
fig,ax = plt.subplots(2,2)
ax[0,0].bar(x=data['Id'] , y=data['SepalWidthCm'])
ax[0,1].bar(x=data['Id'] ,y=data['SepalLengthCm'])
ax[1,0].bar(x=data['Id'] , y=data['PetalWidthCm'])
ax[1,1].bar(x=data['Id'] , y=data['PetalLengthCm'])
plt.show()
```
