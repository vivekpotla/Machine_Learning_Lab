# EXPLORATORY DATA ANALYSIS

Exploratory Data Analysis (EDA) is an approach to analyze the data using visual techniques. It is used to discover trends, patterns, or to check assumptions with the help of statistical summary and graphical representations.

Checking the number of null values of each columns

Checking total number of null values

```python
data_set.head(100).isnull().sum().sum()
```

Filling NULL values with 0 for a Column

```python
data_set['SepalLengthCm']=data_set['SepalLengthCm'].fillna(0)
```

Checking if still NULL are left in SepalLengthCm column

```python
data_set['SepalLengthCm'].head(100).isnull().sum()
```

Checking NULL values in SepalWidthCm column

```python
data_set['SepalWidthCm'].head(100).isnull().sum()
```

Replacing the NULL values in SepalWidthCm column with mean

```python
data_set['SepalWidthCm']=data_set['SepalWidthCm'].fillna(data_set['SepalWidthCm'].mean())
```

Checking if NULL values are replaced or not

```python
data_set['SepalWidthCm'].head(100).isnull().sum()
```

Checking NULL values in PetalLength column

```python
data_set['PetalLengthCm'].head(100).isnull().sum()
```

Replacing NULLvalues in PetalLengthCm with Median value

```python
data_set['PetalLengthCm']=data_set['PetalLengthCm'].fillna(data_set['PetalLengthCm'].median())
```

Checking if NULL values are replaced or not

```python
data_set['PetalLengthCm'].head(100).isnull().sum()
```

CHECKING IF NULL VALUES ARE REPLACED IN THE ENTIRE DATASET

```python
data_set.isnull().sum()
```

```python
print("The Total NULL values are :" ,data_set.isnull().sum().sum())
```
